```python
def url_title_tmdb(row, col):
    urls = []
    if pd.isnull(row[col]):
        title = row['originalTitle']
        # Percent encode the title
        title_encoded = urllib.parse.quote(title, safe='')
        url = f"https://api.themoviedb.org/3/search/movie?query={title_encoded}&include_adult=false&language=en-US&page=1"
        urls.append(url)
    return urls
```

```python
def get_movie_details(url, headers):
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  
        details = response.json()
        return details      
    except requests.exceptions.RequestException as e:
        print(f"Error occurred: {e}")
        return None
```

```python
def fill_missing_start_year(df, headers):
    for index, row in df.iterrows():
        imdb_id = row['tconst']
        url = f"https://api.themoviedb.org/3/find/{imdb_id}?external_source=imdb_id"
        print(url)
        try:
            response = requests.get(url, headers=headers)
            response.raise_for_status()  
            details = response.json()
            movie_results = details.get('movie_results', [])
            if movie_results:
                release_date = movie_results[0].get('release_date')
                if release_date:
                    print(f"Release date for movie {row['originalTitle']}: {release_date}")
                    df.at[index, 'startYear'] = datetime.strptime(release_date, '%Y-%m-%d').year
                    print(f"Updated startYear for: {row['originalTitle']} as {datetime.strptime(release_date, '%Y-%m-%d').year}")
                else:
                    print(f"No release date found for movie: {row['originalTitle']}")
            else:
                print(f"No movie details found for movie: {row['originalTitle']}")
        except requests.exceptions.RequestException as e:
            print(f"Error occurred: {e}")
    
        delay = (random.randint(0,500)/1000)  
        time.sleep(delay)
        print(f"Waiting {delay} seconds...")
```

```python
def fill_missing_data(df, headers):
    for index, row in df.iterrows():
        imdb_id = row['tconst']
        url = f"https://api.themoviedb.org/3/find/{imdb_id}?external_source=imdb_id"
        try:
            response = requests.get(url, headers=headers)
            response.raise_for_status()  
            details = response.json()
            movie_results = details.get('movie_results', [])
            
            if movie_results:
                tmdb_id = movie_results[0]['id']
                api_url = f"https://api.themoviedb.org/3/movie/{tmdb_id}?language=en-US"
                
                try:
                    response2 = requests.get(api_url, headers=headers)
                    response2.raise_for_status()
                    data = response2.json()
                    
                    if df.at[index, 'tconst'] == data['imdb_id']:
                        df.at[index, 'averageRating'] = data['vote_average']
                        df.at[index, 'numVotes'] = data['vote_count']
                        df.at[index, 'primaryTitle'] = data['title']
                        df.at[index, 'originalTitle'] = data['original_title']
                        df.at[index, 'runtimeMinutes'] = data['runtime']
                        df.at[index, 'isAdult'] = data['adult']
                        if 'release_date' in data and data['release_date']:
                            df.at[index, 'startYear'] = int(data['release_date'].split('-')[0])
                        if 'genres' in data and data['genres']:
                            genres_set = set(genre['name'] for genre in data['genres'])
                            df.at[index, 'genres'] = genres_set
                        print(f'Data added for {tmdb_id}')
                except Exception as e:
                    print(f"Error processing URL {api_url}: {e}")
            else:
                df = df[df.index != index]

        except requests.exceptions.RequestException as e:
            print(f"Error occurred: {e}")
                
        time.sleep(0.02)
    return df
```

```python
def format_set(df, col):
    if not isinstance(df[col].iloc[0], set):
        df[col] = df[col].str.split(',')
        df[col] = df[col].apply(lambda x: set(x) if isinstance(x, list) else set())
    else:
        df[col] = df[col]
    return df
```

```python
def get_actors(data, max_order=5, separator=' '):
    actors = []
    for actor in data:
        if actor['order'] <= max_order:
            name = actor['name']
            actors.append(name)
    if actors:
        return separator.join(actors)
```

```python
def get_crew(data, members, separator=' '):
    crew = []
    for member in data:
        if member['job'] in members and member['name'] not in crew:
            name = member['name']
            crew.append(name)
    if crew:
        return separator.join(crew)
```

```python
def get_items(row, col, separator=' '):
    try:
        data = json.loads(row[col])
        items = [item['name'] for item in data]  # Assuming 'name' is the key you want to extract
    except (KeyError, json.JSONDecodeError, TypeError):
        items = []  # Return an empty list if value is not a valid JSON string or column does not exist
    
    return separator.join(items)
```

```python
a['genres'] = a.apply(lambda row: get_items(row, 'genres'), axis=1)
a['keywords'] = a.apply(lambda row: get_items(row, 'keywords'), axis=1)
a['production_countries'] = a.apply(lambda row: get_items(row, 'production_countries'), axis=1)
a['production_companies'] = a.apply(lambda row: get_items(row, 'production_companies'), axis=1)
```
