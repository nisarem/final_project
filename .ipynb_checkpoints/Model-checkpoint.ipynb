```python
def format_name(name):
    if name is None:
        return None
    name = re.sub(r'[^\w\s]', '', name)
    parts = name.split()
    return '_'.join(parts)
```

```python
def get_genres(row):
    if row['genres'] is None:
        return ''
    data = json.loads(row['genres'])
    genres = [format_name(genre['name']) for genre in data]
    
    return genres
```

```python
def get_genres(row):
    if isinstance(row['genres'], list):
        # If genres is already a list of dictionaries
        genres = [format_name(genre['name']) for genre in row['genres']]
    elif isinstance(row['genres'], str):
        # If genres is a string, load it using json.loads
        data = json.loads(row['genres'])
        genres = [format_name(genre['name']) for genre in data]
    else:
        genres = []
    
    return genres
```

```python
def get_keywords(row):
    if isinstance(row['keywords'], list):  # Check if it's already a list
        kws = [kw['name'] for kw in row['keywords']]
        return ' '.join(kws)
    elif isinstance(row['keywords'], str):  # If it's a string, load JSON
        data = json.loads(row['keywords'])
        kws = [kw['name'] for kw in data]
        return ' '.join(kws)
    else:
        return ''
```

```python
def get_actors(data, max_order=5, separator=' '):
    actors = []
    for actor in data:
        if actor['order'] <= max_order:
            name = format_name(actor['name'])
            actors.append(name)
    if actors:
        return separator.join(actors)
    else:
        return ''
```

```python
def get_crew(data, members, separator=' '):
    crew = []
    for member in data:
        if member['job'] in members and format_name(member['name']) not in crew:
            name = format_name(member['name'])
            crew.append(name)
    if crew:
        return separator.join(crew)
    else:
        return ''
```

```python
def get_countries(data, separator=' '):
    top_countries = ['US', 'GB', 'CN', 'FR', 'JP', 'DE', 'KR', 'CA', 'AU', 'IN']

    countries = []
    for country in data:
        name = format_name(country['name'])
        if name not in top_countries:
            countries.append(name)
        
    return separator.join(countries)
```

```python
def get_description(data, num_actors=10):
    # Define keywords, actors, crew, overview, and countries
    keywords = get_keywords(data)
    if isinstance(data['cast'], list):
        actors = get_actors(data['cast'], num_actors)
    elif isinstance(data['cast'], str):
        actors = get_actors(json.loads(data['cast']), num_actors)
    
    if isinstance(data['crew'], list):
        crew = get_crew(data['crew'], ['Producer', 'Director'])
    else:
        crew = get_crew(json.loads(data['crew']), ['Producer', 'Director'])
        
    if isinstance(data['production_countries'], list):
        countries = get_countries(data['production_countries'])
    else:
        countries = get_countries(json.loads(data['production_countries']))
    
    overview = re.sub(r'[^\w\s]', '', data['overview'])


    # Initialize stemmer and lemmatizer
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()

    # Tokenize and process overview
    tokens = word_tokenize(overview)
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

    # Join processed tokens into single strings
    processed_overview_stemmed = ' '.join(stemmed_tokens)
    processed_overview_lemmatized = ' '.join(lemmatized_tokens)

    # Concatenate all components into the description
    description = f"{keywords} {actors} {crew} {processed_overview_stemmed} {countries} {countries} {countries}"
    
    return description
```

```python
def process_release_date(row):
    date = row['release_date']
    if pd.isnull(date):
        return datetime.now().year
    elif isinstance(date, str):
        if '-' in date:
            return int(date.split('-')[0])
        elif date.isdigit() and len(date) == 4:
            return int(date)
        else:
            raise ValueError("Invalid release date format.")
    elif isinstance(date, int):
        return date
    else:
        raise ValueError("Invalid release date format.")
```

```python
def movie_recommender_weighted(movie_id, data, similarity_matrix, genre_matrix, top_n=3, genre_weight=0.4):
    
    movie_index = data.index[data['id'] == movie_id].tolist()[0]
    similarity_scores = similarity_matrix[movie_index]
    genre_scores = genre_matrix[movie_index]

    combined_scores = (1 - genre_weight) * similarity_scores + genre_weight * genre_scores

    indices = np.argsort(combined_scores)[::-1]
    indices = indices[indices != movie_index]
    
    top_indices = indices[:top_n]
    similar_movies = data.iloc[top_indices]['title']
    
    return similar_movies.tolist()
```

```python
def fetch_movie_details(series, header):
    df_combined = pd.DataFrame()  
    for movie_id in series:
        url = f"https://api.themoviedb.org/3/movie/{movie_id}?&append_to_response=keywords,credits"
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            movie_df = format_movie_details_tmdb(response)
            df_combined = pd.concat([df_combined, movie_df], ignore_index=True)  
            
        else:
            print(f"Failed to fetch details for movie with ID: {movie_id}")
            
    data = df_combined[['id','genres','original_title','keywords','overview',
           'popularity','release_date','runtime','vote_average','vote_count',
           'cast','crew','poster_path','production_countries','production_companies']].copy()
    data.rename(columns={'original_title': 'title'}, inplace=True)
    return data
```

```python
def format_movie_details_tmdb(response):
    data = response.json()
    relevant_data = {
        'budget': data['budget'],
        'genres': data['genres'],
        'homepage': data['homepage'],
        'id': data['id'],
        'keywords': data['keywords']['keywords'],
        'original_language': data['original_language'],
        'original_title': data['original_title'],
        'overview': data['overview'],
        'poster_path': data['poster_path'],
        'popularity': data['popularity'],
        'production_countries': data['production_countries'],
        'production_companies': data['production_companies'],
        'release_date': data['release_date'],
        'runtime': data['runtime'],
        'revenue': data['revenue'],
        'spoken_languages': data['spoken_languages'],
        'status': data['status'],
        'tagline': data['tagline'],
        'vote_average': data['vote_average'],
        'vote_count': data['vote_count'],
        'title': data['title'],
        'cast': data['credits']['cast'],
        'crew': data['credits']['crew']
    }

    df = pd.DataFrame([relevant_data])
    return df
```

```python
def preprocess_data(data):

    required_cols = ['genres', 'keywords', 'popularity', 'release_date', 'vote_average', 'cast', 'crew']
    for col in required_cols:
        if col not in data.columns:
            raise ValueError(f"Column '{col}' is missing in the DataFrame.")

    descriptions = []
    for index, row in data.iterrows():
        description = get_description(row)
        descriptions.append(description)
        
    tfidf_scaling_factor = 3
    vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7, max_features=10000)
    tfidf_matrix = vectorizer.fit_transform(descriptions)
    tfidf_matrix_scaled = tfidf_matrix * tfidf_scaling_factor
    tfidf = normalize(tfidf_matrix_scaled)

    scaler = MinMaxScaler()
    release_years = [process_release_date(row) for _, row in data.iterrows()]
    release_year_norm = scaler.fit_transform(np.array(release_years).reshape(-1, 1))
    ratings_norm = scaler.fit_transform(data['vote_average'].values.reshape(-1, 1))
    popularity_norm = scaler.fit_transform(data['popularity'].values.reshape(-1, 1))
    
    genres = []
    for index, row in data.iterrows():
        genre = get_genres(row)
        genres.append(genre) 

    #one hot encoder does not work on list
    mlb = MultiLabelBinarizer()
    genres_mlb = mlb.fit_transform(genres)
    family_index = np.where(mlb.classes_ == "Family")[0]

    # Multiply the encoding of "family" genre by 2
    if family_index.size > 0:
        genres_mlb[:, family_index] *= 2

    combined = np.hstack((tfidf.toarray(), popularity_norm, release_year_norm, ratings_norm))
    similarity_matrix = cosine_similarity(combined)
    genres_matrix = cosine_similarity(genres_mlb)
    np.fill_diagonal(genres_matrix, 0)

    return similarity_matrix, genres_matrix
```

```python
def final_recommender(data):
    title = input('Enter a movie name: ')
    url = f"https://api.themoviedb.org/3/search/movie?query={title}&include_adult=false&language=en-US&page=1"
    res = requests.get(url, headers=headers).json()
    if res and 'results' in res and len(res['results']) > 0:
        release_year = res['results'][0]['release_date'].split('-')[0]
        print('The movie entered \n')
        print(res['results'][0]['title'],'-',release_year)
        print('\nSearching database...')
        
        movie_id = res['results'][0]['id']
        if movie_id in data['id'].values:
            similarity_matrix, genre_matrix = preprocess_data(data)
            recommended = movie_recommender_weighted(movie_id, data, 
                                                     similarity_matrix, genre_matrix,
                                                     top_n=3, genre_weight=0.1)

            print(f"\nMovies similar to '{res['results'][0]['title']}':")
            
            for i, title in enumerate(recommended, start=1):
                print(f"{i}. {title}")
                rec_id = data[data['title'] == title]['id'].values[0]
                rec_res = fetch_movie_details([rec_id],headers)
                if not rec_res.empty: 
                    poster_url = 'https://image.tmdb.org/t/p/original'+ rec_res.iloc[0]['poster_path'] 
                    display(Image(url=poster_url, width=200, height=300))
                lttx = 'https://letterboxd.com/tmdb/'+str(rec_id)
                print(f'Most popular reviews on Letterboxd about {title}\n')
                reviews = get_reviews_from_link(lttx, num_reviews=3)
                for review in reviews:
                    print(review)
                    print('\n')
                    time.sleep(1)
                    

        else:
            time.sleep(1)
            print('\nMore calculations needed... %^&+*?!')
            movie_det = fetch_movie_details({res['results'][0]['id']},headers)
            df_combined = pd.concat([movie_det, data], ignore_index=True)
            similarity_matrix, genre_matrix = preprocess_data(df_combined)
            recommended = movie_recommender_weighted(movie_id, df_combined, 
                                                     similarity_matrix, genre_matrix,
                                                     top_n=3, genre_weight=0.1)
            
            print(f"\nMovies similar to '{res['results'][0]['title']}':")
            
            for i, title in enumerate(recommended, start=1):
                print(f"{i}. {title}")
                rec_id = data[data['title'] == title]['id'].values[0]
                rec_res = fetch_movie_details([rec_id],headers)
                if not rec_res.empty: 
                    poster_url = 'https://image.tmdb.org/t/p/original'+ rec_res.iloc[0]['poster_path'] 
                    display(Image(url=poster_url, width=200, height=300))
                time.sleep(0.5)
                lttx = 'https://letterboxd.com/tmdb/'+str(rec_id)
                print(f'Most popular reviews on Letterboxd about {title}\n')
                reviews = get_reviews_from_link(lttx, num_reviews=3)
                for review in reviews:
                    print(review)
                    print('\n')
                    time.sleep(1)

    else: 
        print(f"We couldn't find {title} in the database. Please check if there is a spelling mistake.")
```
