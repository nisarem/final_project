```python
import gzip
import pandas as pd
import pickle
import requests
import urllib.parse
import random
import time
from datetime import datetime
import numpy as np

from omdbapi.movie_search import GetMovie
```

```python
f = open("token_TMDB.txt","r")
token = f.read()
headers = {"accept": "application/json", "Authorization": f"Bearer {token}"}
```

```python
people = pickle.load(open('name.basics.pkl', 'rb'))
```

```python
people.isna().sum()
```

```python
null_names = people[people['primaryName'].isna()]
```

```python
null_names
```

```python
essai = "https://api.themoviedb.org/3/find/nm0000138?external_source=imdb_id"
res = requests.get(essai, headers=headers)
info = res.json()
display(info)
```

```python
for index, row in null_names.iterrows():
    imdb_id = row['nconst']  # Replace 'nconst' with the correct column name
    url = f"https://api.themoviedb.org/3/find/{imdb_id}?external_source=imdb_id"
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  
        details = response.json()
        print(details)
    except requests.exceptions.RequestException as e:
        print(f"Error occurred while fetching data for IMDb ID {imdb_id}: {e}")
```

```python
people.drop(null_names.index, inplace=True)
```

```python
people['primaryProfession'] = people['primaryProfession'].apply(lambda x: 
                                                                set(x.split(',')) if isinstance(x, str) else x)
```

```python
people['knownForTitles'] = people['knownForTitles'].apply(lambda x: 
                                                          set(x.split(',')) if isinstance(x, str) else x)
```

```python
people
```

```python
people[(people['primaryProfession'].isna())
       &(people['knownForTitles'].isna())
       &(people['deathYear'].isna())
       &(people['birthYear'].isna())]
```

```python
people.dropna(subset=list(people.columns.difference(['nconst', 'primaryName'])), how='all', inplace=True)
```

```python
people.dropna(subset=['primaryProfession', 'knownForTitles'], how='all', inplace=True)
```

```python
people['primaryProfession'].explode().unique()
```

```python
valid_professions = ['actor', 'actress', 'director']
# Filter rows where primaryProfession is either null or one of the valid professions
filtered_people = people[people['primaryProfession'].isna() | 
                         people['primaryProfession'].apply(lambda x: isinstance(x, set) 
                                                           and bool(x.intersection(valid_professions)))]
```

```python
filtered_people
```

```python
with open('name.basics.filtered.pkl', 'wb') as f:
    pickle.dump(filtered_people, f)
```
