```python
def url_title_tmdb(row, col):
    urls = []
    if pd.isnull(row[col]):
        title = row['originalTitle']
        # Percent encode the title
        title_encoded = urllib.parse.quote(title, safe='')
        url = f"https://api.themoviedb.org/3/search/movie?query={title_encoded}&include_adult=false&language=en-US&page=1"
        urls.append(url)
    return urls
```

```python
def get_name_from_id(person_id, people_df):
    if isinstance(person_id, str):
        person_id = person_id.strip()  # Strip whitespace if it's a string
    if ',' in person_id:
        # Split multiple IDs by comma, strip each ID, and fetch names
        return [get_name_from_id(id.strip(), people_df) for id in person_id.split(',')]
    else:
        person = people_df[people_df['nconst'] == person_id]
        if not person.empty:
            return person.iloc[0]['primaryName']
        else:
            return None
```
