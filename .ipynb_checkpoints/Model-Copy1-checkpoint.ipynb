```python
def get_genres(row):
    if row['genres'] is None:
        return ''
    data = json.loads(row['genres'])
    genres = [genre['name'] for genre in data]
    return ' '.join(genres)
```

```python
def get_keywords(row):
    if row['keywords'] is None:
        return ''
    data = json.loads(row['keywords'])
    kws = [kw['name'] for kw in data]
    return ' '.join(kws)
```

```python
def format_name(name):
    if name is None:
        return None
    name = re.sub(r'[^\w\s]', '', name)
    parts = name.split()
    return '_'.join(parts)
```

```python
def get_actors(data, max_order=10, separator=' '):
    actors = []
    for actor in data:
        if actor['order'] <= max_order:
            name = format_name(actor['name'])
            actors.append(name)
    if actors:
        return separator.join(actors)
    else:
        return ''
```

```python
def get_crew(data, members, separator=' '):
    crew = []
    for member in data:
        if member['job'] in members and format_name(member['name']) not in crew:
            name = format_name(member['name'])
            crew.append(name)
    if crew:
        return separator.join(crew)
    else:
        return ''
```

```python
def country_to_continent(country_name):
    try:
        country_alpha2 = pc.country_name_to_country_alpha2(country_name)
        country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
        country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
        return country_continent_name
    except Exception as e:
        return ''
```

```python
def get_countries(data, separator=' '):
    # Define top countries
    top_countries = ['US', 'GB', 'CN', 'FR', 'JP', 'DE', 'KR', 'CA', 'AU', 'IN']

    countries = []
    for country in data:
        if 'iso_3166_1' in country:
            iso_code = country['iso_3166_1']
            if iso_code in top_countries:
                name = format_name(country['name'])
                countries.append(name)
            else:
                continent = country_to_continent(country['name'])
                if continent is not None:  # Check if continent is not None
                    continent = format_name(continent)
                    countries.append(continent)
    if countries:
        return separator.join(countries)
    else:
        return ''
```

```python
def get_companies(data, separator=' '):
    companies = []
    for company in data:
        if company['name']:
            company = format_name(company['name'])
            companies.append(company)
    if companies:
        return separator.join(companies)
    else:
        return ''
```

```python
def get_description(data, num_actors=10):
    keywords = get_keywords(data)
    if keywords is None:
        keywords = ''

    actors = get_actors(json.loads(data['cast']), num_actors)
    if actors is None:
        actors = ''

    crew = get_crew(json.loads(data['crew']), ['Producer', 'Director'])
    if crew is None:
        crew = ''

    overview = data['overview']
    if overview is None:
        overview = ''

    description = f"{keywords} {actors} {crew} {overview}"
    
    return description
```

```python
def get_production(data):
    d = json.loads(data['production_countries'])
    e = json.loads(data['production_companies'])
    countries = get_countries(d)
    companies = get_companies(e)
    return countries, companies
```

```python
productions_countries = []
productions_companies = []

for index, row in data.iterrows():
    countries, companies = get_production(row)
    productions_countries.append(countries)
    productions_companies.append(companies)
```

```python
unique_countries = set()
unique_companies = set()
for countries, companies in zip(productions_countries, productions_companies):
    if countries:
        unique_countries.update(countries.split())
    if companies:
        unique_companies.update(companies.split())

mlb_countries = MultiLabelBinarizer(classes=sorted(unique_countries))
mlb_companies = MultiLabelBinarizer(classes=sorted(unique_companies))
countries_enc = mlb_countries.fit_transform([countries.split() if countries else [] for countries in productions_countries])
companies_enc = mlb_companies.fit_transform([companies.split() if companies else [] for companies in productions_companies])
production_enc = np.hstack((countries_enc, companies_enc))
```

```python
def movie_recommender_weighted(movie_id, data, similarity_matrix, genre_matrix, production_similarity_matrix, top_n=3, genre_weight=0.3, production_weight=0.2):
    
    movie_index = data.index[data['id'] == movie_id].tolist()[0]
    similarity_scores = similarity_matrix[movie_index]
    genre_scores = genre_matrix[movie_index]
    production_scores = production_similarity_matrix[movie_index]

    combined_scores = (1 - genre_weight - production_weight) * similarity_scores + genre_weight * genre_scores + production_weight * production_scores

    indices = np.argsort(combined_scores)[::-1]
    indices = indices[indices != movie_index]
    
    top_indices = indices[:top_n]
    similar_movies = data.iloc[top_indices]['title']
    
    return similar_movies.tolist()
```

```python
def preprocess_data(data):

    required_cols = ['genres', 'keywords', 'popularity', 'release_date', 'vote_average', 'cast', 'crew']
    for col in required_cols:
        if col not in data.columns:
            raise ValueError(f"Column '{col}' is missing in the DataFrame.")

    if data['cast'].apply(lambda x: len(x) == 0).any():
        raise ValueError("Some rows in the 'cast' column have no actors.")
    if data['crew'].apply(lambda x: len(x) == 0).any():
        raise ValueError("Some rows in the 'crew' column have no members.")

    descriptions = []
    for index, row in data.iterrows():
        description = get_description(row)
        descriptions.append(description)

    vectorizer = TfidfVectorizer()
    tfidf = vectorizer.fit_transform(descriptions)

    scaler = MinMaxScaler()
    popularity_norm = scaler.fit_transform(data['popularity'].values.reshape(-1, 1))
    release_year_norm = scaler.fit_transform(data['release_date'].str[:4].astype(float).values.reshape(-1, 1))
    ratings_norm = scaler.fit_transform(data['vote_average'].values.reshape(-1, 1))

    combined = np.hstack((tfidf.toarray(), popularity_norm, release_year_norm, ratings_norm))
    similarity_matrix = cosine_similarity(combined)
    
    genres = []
    for index, row in data.iterrows():
        genre = get_genres(row)
        genres.append(genre)
    
    mlb = MultiLabelBinarizer()
    genre_encoded = mlb.fit_transform(genres)
    genre_matrix = cosine_similarity(genre_encoded)
    np.fill_diagonal(genre_matrix, 0)

    return similarity_matrix, genre_matrix
```

```python
similarity_matrix, genre_matrix = preprocess_data(data)
movie_recommender_weighted(116149, data, similarity_matrix, genre_matrix)
```
